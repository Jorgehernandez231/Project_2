```python
import pandas as pd
```

```python
dt = pd.read_csv('happiness_all_years_clean.csv')
country_id = pd.read_csv('country_id.csv')
dt
```

```python
country_id
```

```python
dt1 = pd.merge(
    dt,
    country_id[['country', 'country_id']],
    on='country',
    how='left'
)
```

```python
dt1
```

```python
dt1 = dt1.drop(columns=['country'])
dt1
```

```python
dt1 = dt1.drop(columns=['region'])
```

```python
col = ['country_id','year', 'score', 'gdp_per_capita', 'social_support','healthy_life_expectancy', 'freedom', 'generosity', 'corruption']

dt1 = dt1[col]
dt1
```
