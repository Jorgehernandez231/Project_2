```python


years = [2015, 2016, 2017, 2018, 2019]
paths = {
    2015: 'data_2015.csv',
    2016: 'data_2016.csv',
    2017: 'data_2017.csv',
    2018: 'data_2018.csv',
    2019: 'data_2019.csv'
}

dfs = []
for y in years:
    df = pd.read_csv(paths[y])
    df['year'] = y
    dfs.append(df)
```

```python
# Read one file with 'Region' info (e.g., 2015)
df_2015 = pd.read_csv(paths[2015])
if 'Region' in df_2015.columns:
    region_map = df_2015[['Country', 'Region']].drop_duplicates().set_index('Country')['Region'].to_dict()
else:
    region_map = {}

# If you have 'Region' in 2016/2017, you can expand the mapping:
for year in [2016, 2017]:
    df = pd.read_csv(paths[year])
    if 'Region' in df.columns:
        region_map.update(df[['Country', 'Region']].drop_duplicates().set_index('Country')['Region'].to_dict())
```

```python
for df in dfs:
    df['region'] = df['Country'].map(region_map)
```

```python
col_rename = {
    'Country': 'country',
    'Happiness Score': 'score',
    'GDP per capita': 'gdp_per_capita',
    'Social support': 'social_support',
    'Healthy life expectancy': 'healthy_life_expectancy',
    'Freedom': 'freedom',
    'Generosity': 'generosity',
    'Perceptions of corruption': 'corruption',
    'region': 'region',
    'year': 'year'
}

for i in range(len(dfs)):
    dfs[i] = dfs[i].rename(columns=col_rename)
```

```python
# Concatenate all DataFrames
df_all = pd.concat(dfs, ignore_index=True)

# Optional: Check missing regions
missing_regions = df_all[df_all['region'].isnull()]['country'].unique()
print("Countries missing region info:", missing_regions)

# Export to CSV for MySQL import
df_all.to_csv("happiness_all_years_clean.csv", index=False)
```

```python
# Manually update missing regions
manual_regions = {
    'Taiwan Province of China': 'Eastern Asia',
    'Hong Kong S.A.R., China': 'Eastern Asia',
    'Trinidad & Tobago': 'Latin America and Caribbean',
    'Northern Cyprus': 'Western Asia',
    'North Macedonia': 'Central and Eastern Europe',
    'Gambia': 'Sub-Saharan Africa'
}

# Fill missing regions
df_all['region'] = df_all.apply(
    lambda row: manual_regions.get(row['country'], row['region']),
    axis=1
)

# Optional: Re-check for any missing regions
still_missing = df_all[df_all['region'].isnull()]['country'].unique()
print("Still missing regions:", still_missing)
```

```python
df_all.to_csv("happiness_all_years_clean.csv", index=False)
```
